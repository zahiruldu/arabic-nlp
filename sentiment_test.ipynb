```python
import matplotlib.pyplot as plt
from camel_tools.sentiment import SentimentAnalyzer

# Initialize sentiment analyzer
sa = SentimentAnalyzer.pretrained()

# Arabic sentences for sentiment analysis
sentences = [
    'مال الهوى و مالي شكون اللي جابني ليك ما كنت انايا ف حالي بلاك',
    'بانا بيك بيك دب قلبي يجبدن بحبك انا مجنون ما ينسى حبك يوم',
    'انا مجنون ما ينسى حبك'
]

# Perform sentiment analysis
sentiments = sa.predict(sentences)

# Plot pie chart for each sentence
plt.figure(figsize=(10, 5))

for i, (sentence, sentiment) in enumerate(zip(sentences, sentiments)):
    plt.subplot(1, len(sentences), i + 1)  # Create subplot for each sentence
    sizes = [1]  # Full pie for each sentiment
    color = 'green' if sentiment == 'positive' else 'orange'
    plt.pie(sizes, labels=[sentiment], colors=[color], startangle=90)
    plt.title(f"Sentence {i+1} Sentiment")

plt.suptitle("Sentiment Analysis of Arabic Poetry (Per Sentence)")
plt.show()
```
